### Build a relationship between user and personality within PDB database

```python
import requests

def get_types(realname):
    url = f"https://api.personality-database.com/api/v1/search?q={realname}"
    payload={"query": realname}
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_0_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}
    response = requests.request("POST", url, headers=headers, data=payload)
    if response.status_code != 200:
        return ''
    data = response.json()
    return data[0]['personality_type'][:4]
```
